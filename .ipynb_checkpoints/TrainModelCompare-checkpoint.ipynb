# Feed Forward Neural Network (FFNN)

### Split dataset

For this network architecture we will split dataset into:

**Train set**

**validation set**

**Test set**

```python
def split_num_ratio(x,q1,q2,q3):
    p1 = x*q1
    p2 = p1+(x*q2)
    p3 = p2+(x*q3)
    return p1,p2,p3

def index_split(data,q1,q2,q3):
    maxlen = len(data)
    p1,p2,p3 = split_num_ratio(maxlen,q1,q2,q3)

    ind1 = round(p1)-1
    ind2 = round(p2)-1
    ind3 = round(p3)-1
    return ind1,ind2,ind3

def data_split(data,q1,q2,q3):
    """
        Returns splitted dataset 
    """
    ind1,ind2,ind3 = index_split(data,q1,q2,q3)
    
    d1 = data[0:ind1]
    d2 = data[ind1+1:ind2]
    d3 = data[ind2+1:ind3]
    return d1,d2,d3
```

```python
inp = np.asarray(df_good_inp)
tar_vm1x = np.asarray(df_good_target['VM1X'])
tar_vm1y = np.asarray(df_good_target['VM1Y'])
tar_vm2x = np.asarray(df_good_target['VM2X'])
tar_vm2y = np.asarray(df_good_target['VM2X'])

train_input, val_input, test_input = data_split(inp,.60,.20,.20)

train_target_vm1x,val_target_vm1x,test_target_vm1x = data_split(tar_vm1x,.60,.20,.20)

train_target_vm1y,val_target_vm1y,test_target_vm1y = data_split(tar_vm1y,.60,.20,.20)

train_target_vm2x,val_target_vm2x,test_target_vm2x = data_split(tar_vm2x,.60,.20,.20)

train_target_vm2y,val_target_vm2y,test_target_vm2y = data_split(tar_vm2y,.60,.20,.20)
```

### Instantiate and Train Model

```python
x = NNM.TFNNModels()
model_vm1x = x.DNN_FFSeqSGD_Regulized(Learn_Rate_Schedule = False)
model_vm1y = x.DNN_FFSeqSGD_Regulized(Learn_Rate_Schedule = False)
model_vm2x = x.DNN_FFSeqSGD_Regulized(Learn_Rate_Schedule = False)
model_vm2y = x.DNN_FFSeqSGD_Regulized(Learn_Rate_Schedule = False)
```

```python
history1 = model_vm1x.fit(train_input,train_target_vm1x,validation_data=(val_input, val_target_vm1x),
epochs=50, batch_size=10)
```

```python
model_vm1x.save('Models\FFmodel_vm1x') # Save Model
```

```python
history2 = model_vm1y.fit(train_input,train_target_vm1y,validation_data=(val_input, val_target_vm1y),
epochs=50, batch_size=10)
```

```python
model_vm1y.save('Models\FFmodel_vm1y') # Save Model
```

```python
history3 = model_vm2x.fit(train_input,train_target_vm2x,validation_data=(val_input, val_target_vm2x),
epochs=50, batch_size=10)
```

```python
model_vm2x.save('Models\FFmodel_vm2x') # Save Model
```

```python
history4 = model_vm2y.fit(train_input,train_target_vm2y,validation_data=(val_input, val_target_vm2y),
epochs=50, batch_size=10)
```

```python
model_vm2y.save('Models\FFmodel_vm2y') # Save Model
```

## Evaluate

```python
import matplotlib.pyplot as plt
def plot_history(history):
  plt.figure()
  plt.xlabel('Epoch')
  plt.ylabel('Mean Abs Error [1000$]')
  plt.plot(history.epoch, np.array(history.history['mae']), 
           label='Train')
  plt.plot(history.epoch, np.array(history.history['val_mae']),
           label = 'Val')
  plt.legend()
  plt.ylim([0, np.max(history.history['val_mae'])+ .30*np.max(history.history['val_mae'])])
  
def plot_prediction(test_labels, test_predictions):
  plt.figure()
  plt.scatter(test_labels, test_predictions)
  plt.xlabel('True Values [1000$]')
  plt.ylabel('Predictions [1000$]')
  plt.axis('equal')
  plt.xlim(plt.xlim())
  plt.ylim(plt.ylim())
  _ = plt.plot([-100, 100],[-100,100])

  plt.figure()
  error = test_predictions - test_labels
  plt.hist(error, bins = 50)
  plt.xlabel("Prediction Error [1000$]")
  _ = plt.ylabel("Count")
```

# Convolutional Neural Network (CNN)

### Data PreProcess

```python
df_good_inp = pd.read_csv('DataFiles/new_inp.csv')  # import data
df_good_target = pd.read_csv('DataFiles/new_tar.csv')  # import data
```

```python
# Train and Test input (70-30 train test split)
splitX = round(.70*len(df_good_inp))
max = len(df_good_inp)

trainX = df_good_inp[0:splitX]
testX = df_good_inp[splitX+1:max]
```

```python
# Train and Test target (70-30 train test split)
splitY = round(.70*len(df_good_target))
max = len(df_good_target)

trainY = df_good_target[0:splitY]
testY = df_good_target[splitY+1:max]
```

```python
# input data preprocessing
trainX = np.asarray(trainX)
testX = np.asarray(testX)

sample_size = trainX.shape[0] # number of samples in train set
time_steps  = trainX.shape[1] # number of features in train set
input_dimension = 1               # each feature is represented by 1 number

train_data_reshaped = trainX.reshape(sample_size,time_steps,input_dimension)
test_data_reshaped = testX.reshape(testX.shape[0],testX.shape[1],1)
```

```python
# target data preprocessing
train_target1 = np.asarray(trainY['VM1X'])
test_target1 = np.asarray(testY['VM1X'])

train_target2 = np.asarray(trainY['VM1Y'])
test_target2 = np.asarray(testY['VM1Y'])

train_target3 = np.asarray(trainY['VM2X'])
test_target3 = np.asarray(testY['VM2X'])

train_target4 = np.asarray(trainY['VM2Y'])
test_target4 = np.asarray(testY['VM2Y'])
```

### CNN Model

```python
# param
kernel_size = 2
n_timesteps = train_data_reshaped.shape[1] #13
n_features  = train_data_reshaped.shape[2] #1 
```

```python
NNM = NNM.TFNNModels()
model_conv1D_tar1 = NNM.build_conv1D_model(n_timesteps,n_features,kernel_size)
model_conv1D_tar2 = NNM.build_conv1D_model(n_timesteps,n_features,kernel_size)
model_conv1D_tar3 = NNM.build_conv1D_model(n_timesteps,n_features,kernel_size)
model_conv1D_tar4 = NNM.build_conv1D_model(n_timesteps,n_features,kernel_size)
```

### Train Model

```python
# Store training stats
cnn_history1 = model_conv1D_tar1.fit(train_data_reshaped, train_target1, epochs=50,
                    validation_split=0.2, verbose=0)
```

```python
cnn_history2 = model_conv1D_tar2.fit(train_data_reshaped, train_target2, epochs=50,
                    validation_split=0.2, verbose=0)
```

```python
cnn_history3 = model_conv1D_tar3.fit(train_data_reshaped, train_target3, epochs=50,
                    validation_split=0.2, verbose=0)
```

```python
cnn_history4 = model_conv1D_tar4.fit(train_data_reshaped, train_target4, epochs=50,
                    validation_split=0.2, verbose=0)
```

```python
model_conv1D_tar1.save('Models\ Newmodel_conv1D_tar1') # Save Model
model_conv1D_tar2.save('Models\ Newmodel_conv1D_tar2') # Save Model
model_conv1D_tar3.save('Models\ Newmodel_conv1D_tar3') # Save Model
model_conv1D_tar4.save('Models\ Newmodel_conv1D_tar4') # Save Model
```
